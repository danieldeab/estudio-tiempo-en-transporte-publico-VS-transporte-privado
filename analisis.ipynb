```python
import pandas as pd

df = pd.read_csv('combinado1.csv')
```

```python
total_filas = len(df)
completitud = pd.DataFrame({
    'valores_completos': df.notna().sum(),
    'valores_nulos': df.isna().sum(),
    'porcentaje_completo': (df.notna().sum() / total_filas * 100).round(2)
})

print(f'Total filas: {total_filas}\n')
completitud
```

```python
transit = pd.read_csv('transit.csv')
filas_con_nulo = transit[transit.isna().any(axis=1)]
print(f'Filas TRANSIT con al menos un nulo: {len(filas_con_nulo)} de {len(transit)} ({len(filas_con_nulo)/len(transit)*100:.1f}%)')
```
