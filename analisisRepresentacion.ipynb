## Representación gráfica de la duración de los trayectos

En esta sección se representan gráficamente las distribuciones de duración de los viajes para comparar los dos modos de transporte analizados:

- **`DRIVE`**: duración del trayecto realizado en coche.
- **`TRANSIT`**: duración del trayecto realizado en transporte público.

Para ello, las duraciones se agrupan en **intervalos de tiempo (minutos)**:

- **`[0, 10)`**
- **`[10, 20)`**
- **`[20, 30)`**
- **`[30, 45)`**
- **`[45, 60)`**
- **`[60, 90)`**
- **`[90, 120)`**

Cada intervalo contiene la **frecuencia absoluta (`ni`)**, es decir, el **número de trayectos cuya duración se encuentra dentro de ese rango de tiempo**.

### Gráfico de barras

El gráfico de barras permite **comparar directamente la frecuencia de viajes entre ambos modos de transporte en cada intervalo de duración**.

- El **eje X** representa los **intervalos de duración del trayecto (minutos)**.
- El **eje Y** representa la **frecuencia absoluta (`ni`)**, es decir, el número de viajes observados en cada intervalo.
- Para cada intervalo se muestran **dos barras**:
  - una correspondiente a **`DRIVE`**
  - otra correspondiente a **`TRANSIT`**

La **altura de cada barra** indica cuántos trayectos tienen una duración dentro de ese intervalo.

### Gráfico de líneas

El gráfico de líneas utiliza los **mismos datos de frecuencia**, pero conecta los valores mediante líneas para facilitar la observación de la **tendencia en la distribución de las duraciones**.

- Cada **punto** representa la frecuencia de viajes en un intervalo.
- Las líneas permiten visualizar **cómo cambia la frecuencia a medida que aumenta la duración del trayecto**.
- Se representan **dos líneas**, una para **`DRIVE`** y otra para **`TRANSIT`**, lo que permite comparar la evolución de las frecuencias entre ambos modos de transporte.

### Interpretación general

A partir de estas representaciones gráficas se puede observar la **distribución de las duraciones de viaje** según el modo de transporte.

En general:

- Los trayectos en **`DRIVE`** tienden a concentrarse en **intervalos de duración más cortos**, especialmente entre **10 y 30 minutos**.
- Los trayectos en **`TRANSIT`** presentan una mayor frecuencia en **intervalos de duración más largos**, particularmente entre **30 y 60 minutos**.

Esto sugiere que, dentro del conjunto de datos analizado, los viajes realizados en transporte público suelen implicar **mayores tiempos de desplazamiento que los realizados en coche**.

```python
import matplotlib.pyplot as plt
import numpy as np

intervalos = ["[0,10)", "[10,20)", "[20,30)", "[30,45)", "[45,60)", "[60,90)", "[90,120)"]

drive = [39, 462, 435, 85, 24, 49, 22]
transit = [19, 95, 236, 326, 203, 174, 137]

x = np.arange(len(intervalos))
width = 0.35

# -------- GRAFICO DE BARRAS --------
plt.figure()

plt.bar(x - width/2, drive, width, label="DRIVE")
plt.bar(x + width/2, transit, width, label="TRANSIT")

plt.xlabel("Duración del viaje (min)")
plt.ylabel("Frecuencia")
plt.title("Comparación DRIVE vs TRANSIT (Barras)")

plt.xticks(x, intervalos)

max_val = max(max(drive), max(transit))
plt.yticks(np.arange(0, max_val + 50, 50))

plt.legend()
plt.grid(axis="y")

plt.show()


# -------- GRAFICO DE LINEAS --------
plt.figure()

plt.plot(x, drive, marker="o", label="DRIVE")
plt.plot(x, transit, marker="o", label="TRANSIT")

plt.xlabel("Duración del viaje (min)")
plt.ylabel("Frecuencia")
plt.title("Comparación DRIVE vs TRANSIT (Líneas)")

plt.xticks(x, intervalos)
plt.yticks(np.arange(0, max_val + 50, 50))

plt.legend()
plt.grid(True)

plt.show()
```
